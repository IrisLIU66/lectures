## Lists

What is the output of this code?

```python
a = [1, 2, 3] # 把对象存储在一个地方，并贴上标签a
b = a # a,b标签指向同一个地方(它们都只是标签，地址相同)
b[1] = 6 # 通过标签b修改对象，对象的内容发生了改变，但a,b两个标签仍指向它
print(a, b) # Output: [1, 6, 3] [1, 6, 3]
```

```python
# How to acturally separate them ?
a = [1, 2, 3]
b = a.copy() # 这才是复制并创建新的list(地址不同)
b[1] = 6 
print(a, b)
```

## List methods

[List methods - Python documentation](https://docs.python.org/3/tutorial/datastructures.html)

Find the 3 top scores out of a list:

```python
all_scores = [23, 67, 92, 54, 77, 66, 72, 70]
```

```python
sorted_scores = sorted(all_scores) # 生成一个新的sorted list
print(sorted_scores)
print(all_scores) # 原list没变
```

```python
all_scores.sort() # 把原list排序了
print(all_scores)
```

```python
print(all_scores[:3]) # 参数规则回顾：[start:stop:step] 输出到stop-1处停止
```

```python
print(all_scores[:-4:-1]) # 从0开始，到-4处结束，倒向，每一个输出
```

```python
print(all_scores[-3:])
```

```python
sorted_scores = sorted(all_scores, reverse=True)[:3] # reverse=True: 从大到小排序 (默认是从小到大排)
print(sorted_scores)
```
